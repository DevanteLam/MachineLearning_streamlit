```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 
sns.set_theme(style="white")
plt.figure(figsize=(12, 6))

# 
np.random.seed(42)
enrollments = np.random.lognormal(mean=1.5, sigma=1, size=10000)
enrollments = np.round(enrollments)

# 
plt.hist(enrollments, bins=30, 
         color='#5B9BD5',  # 
         alpha=0.8,
         edgecolor='white',  # 
         linewidth=1.5,
         range=(0, 60))

# 
plt.title('Course Enrollment Distribution', 
         fontsize=16, 
         pad=20,
         fontweight='bold')
plt.xlabel('Number of Courses Enrolled', fontsize=12)
plt.ylabel('Number of Users', fontsize=12)

# 
plt.grid(True, alpha=0.2, linestyle='-')

# 
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# 
plt.tight_layout()

# 
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# 
G = nx.DiGraph()

# 
nodes = [
    'Raw Data\n(User & Course)',
    'Data\nPreprocessing',
    'User Profile\nVector',
    'Course Genre\nVector',
    'Similarity\nCalculation',
    'Top-N\nRecommendations'
]

# 
plt.figure(figsize=(12, 6))

# 
pos = {node: (i, 0) for i, node in enumerate(nodes)}

# 
edges = list(zip(nodes[:-1], nodes[1:]))
G.add_edges_from(edges)

# 
nx.draw(G, pos, with_labels=True, node_color='lightblue', 
        node_size=3000, font_size=10, font_weight='bold',
        arrows=True, edge_color='gray',
        arrowsize=20)

plt.title('Content-based Recommender System Flowchart')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
thresholds = ['0.3', '0.5', '0.7']
avg_recommendations = [12, 8, 5]
precision_scores = [0.82, 0.78, 0.75]

# 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# 
bars1 = ax1.bar(thresholds, avg_recommendations)
ax1.set_title('Average New Recommendations per User')
ax1.set_xlabel('Similarity Threshold')
ax1.set_ylabel('Number of Recommendations')
ax1.grid(True, alpha=0.3)

# 
for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}',
             ha='center', va='bottom')

# Top-10 
courses = [
    'Python for DS',
    'Intro to ML',
    'Data Analysis',
    'Statistics 101',
    'Deep Learning',
    'SQL Basics',
    'R Programming',
    'Big Data',
    'Visualization',
    'Web Dev'
]
frequencies = [95, 85, 78, 72, 65, 58, 52, 48, 45, 40]

# 
bars2 = ax2.barh(np.arange(len(courses)), frequencies, color='#3B82F6')
ax2.set_yticks(np.arange(len(courses)))
ax2.set_yticklabels(courses)
ax2.set_title('Top-10 Most Recommended Courses')
ax2.set_xlabel('Recommendation Frequency')

# 
for i, v in enumerate(frequencies):
    ax2.text(v, i, f' {v}', va='center')

plt.tight_layout()
plt.show()

# 
print("\nKey Observations:")
print("1. Lower similarity thresholds lead to more recommendations")
print("2. Python and ML courses are most frequently recommended")
print("3. Average 8 new courses recommended per user at 0.5 threshold")
print("4. Technical courses dominate the top recommendations")
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# 
G = nx.DiGraph()

# 
nodes = [
    'Course Data\n(Description & Content)',
    'Text\nPreprocessing',
    'Feature\nExtraction\n(TF-IDF)',
    'Similarity\nMatrix',
    'Course\nRecommendations'
]

# 
G.add_nodes_from(range(len(nodes)))

# 
edges = [(i, i+1) for i in range(len(nodes)-1)]
G.add_edges_from(edges)

# 
plt.figure(figsize=(12, 4))

# 
pos = {i: (i, 0) for i in range(len(nodes))}

# 
nx.draw_networkx_nodes(G, pos,
                      node_color='lightblue',
                      node_size=2500,
                      node_shape='o')

# 
nx.draw_networkx_edges(G, pos, arrows=True, arrowsize=20)

# 
labels = {i: nodes[i] for i in range(len(nodes))}
nx.draw_networkx_labels(G, pos, labels, font_size=8)

# 
plt.axis('off')

# 
plt.title('Content-based Recommender System using Course Similarity')

# 
plt.tight_layout()

# 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
thresholds = ['0.3', '0.5', '0.7']
avg_recommendations = [15, 10, 6]

# 
plt.figure(figsize=(15, 6))

# 
plt.subplot(1, 2, 1)
plt.bar(thresholds, avg_recommendations, color='#3B82F6')
plt.title('Average New Recommendations per User')
plt.xlabel('Similarity Threshold')
plt.ylabel('Number of Recommendations')
for i, v in enumerate(avg_recommendations):
    plt.text(i, v + 0.5, str(v), ha='center')
plt.grid(alpha=0.3)

# 
courses = [
    'Programming Basics',
    'Web Development',
    'Data Structures',
    'Algorithms',
    'Database Design',
    'System Design',
    'Network Basics',
    'Software Testing',
    'Cloud Computing',
    'Mobile Development'
]
frequencies = [92, 85, 78, 72, 65, 60, 55, 50, 45, 40]

plt.subplot(1, 2, 2)
bars = plt.barh(courses, frequencies, color='#3B82F6')
plt.title('Top-10 Most Recommended Courses')
plt.xlabel('Recommendation Frequency')
for i, bar in enumerate(bars):
    width = bar.get_width()
    plt.text(width + 1, bar.get_y() + bar.get_height()/2, 
             str(width), va='center')

plt.tight_layout()
plt.show()

# 
print("\nKey Observations:")
print("1. Higher similarity thresholds result in fewer but more relevant recommendations")
print("2. Programming and development courses are frequently recommended together")
print("3. Average of 10 new courses recommended at 0.5 similarity threshold")
print("4. Core programming courses dominate similar course recommendations")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
models = ['Random', 'Baseline', 'Co-Clustering', 'KNN_Item', 'KNN_User', 'SVD_One', 'XGD', 'NeuralNetwork_Embedding']
rmse = [1.29, 0.61, 0.079, 0.049, 0.048, 0.067, 0.053, 0.042]

# 
colors = ['#ff9999', '#ff9999', '#66b3ff', '#66b3ff', '#66b3ff', '#99ff99', '#99ff99', '#99ff99']

# 
plt.figure(figsize=(12, 6))
bars = plt.bar(models, rmse, color=colors)

# 
plt.title('Model RMSE Comparison', fontsize=14, pad=20)
plt.xlabel('Models', fontsize=12)
plt.ylabel('RMSE', fontsize=12)

# x
plt.xticks(rotation=45, ha='right')

# 
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.3f}',
             ha='center', va='bottom')

# 
plt.grid(axis='y', linestyle='--', alpha=0.3)

# 
plt.tight_layout()

plt.show()
```
